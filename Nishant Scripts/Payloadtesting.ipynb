```python
import time
import random
import json
from datetime import datetime, timezone
import requests

# Dozmate API endpoint
API_URL = "https://admin.dozemate.com/api/http/ingest"
HEADERS = {"Content-Type": "application/json"}

DEVICE_ID = "0102-ACDE48112233"  # Updated device ID

def generate_test_data():
    return {
        "timestamp": datetime.now(timezone.utc).isoformat(),
        "heartRate": random.randint(60, 100),
        "respiration": random.randint(12, 20),
        "SpO2": random.randint(90, 100),
        "stress": random.randint(10, 30),
        "temperature": round(random.uniform(36.0, 37.5), 1),
        "humidity": random.randint(40, 60),
        "iaq": random.randint(50, 150),
        "eco2": random.randint(400, 700),
        "tvoc": round(random.uniform(0.0, 1.0), 2),
        "etoh": round(random.uniform(0.0, 0.1), 2),
        "metrics": {
            "sdnn": round(random.uniform(30, 60), 2),
            "rmssd": round(random.uniform(20, 50), 2),
            "lf_hf_ratio": round(random.uniform(0.1, 2.0), 2),
            "stress_index": random.randint(10, 30)
        },
        "signals": {
            "motion": random.randint(0, 1),
            "presence": random.randint(0, 1),
            "activity": random.randint(0, 3),
            "battery": random.randint(40, 100),
            "mic": random.randint(0, 1),
            "rrIntervals": [random.randint(700, 900) for _ in range(4)],
            "rawWaveform": [round(random.uniform(0.05, 0.1), 2) for _ in range(4)]
        }
    }

def send_data():
    payload = {
        "deviceId": "0102-ACDE48112233",
        "type": "health",
        "data": generate_test_data()
    }
    response = requests.post(API_URL, headers=HEADERS, json=payload)
    if response.status_code == 200:
        print(f"✅ Data sent successfully: {json.dumps(payload)}")
    else:
        print(f"❌ Failed to send data: {response.status_code}, {response.text}")

# Send data every 6 seconds
try:
    while True:
        send_data()
        time.sleep(6)
except KeyboardInterrupt:
    print("🔌 Stopped sending data.")
```
